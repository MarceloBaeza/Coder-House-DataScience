```python
import requests
import json
import pandas as pd

url = 'https://jsonplaceholder.typicode.com/users'
respuesta = requests.get(url) #PUT, POST

texto = respuesta.text
jsondata = json.loads(texto)

for parada in jsondata: # desapilando Direccion
    Address = parada.pop("address")
    parada["street"] = Address["street"]
    parada["suite"] = Address["suite"]
    parada["city"] = Address["city"]
    parada["zipcode"] = Address["zipcode"]
    parada["geo"] = Address["geo"] 

for parada in jsondata: # desapilando  latitud y longitud
    Address = parada.pop("geo")
    parada["latitud"] = Address["lat"]
    parada["longitud"] = Address["lng"]

for parada in jsondata: # desapilando datos de la compa√±ia
    company = parada.pop("company")
    parada["company name"] = company["name"]
    parada["company catchPhrase"] = company["catchPhrase"]
    parada["company bs"] = company["bs"]

df = pd.DataFrame(jsondata)
df.head()
```
